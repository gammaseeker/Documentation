# LinkedList-Concepts

Linked List: A chain of nodes which points to each subsequent node.
There are 4 flavors of linked lists:
<ul>
<li>Singly Linked List</li>
<li>Doubly Linked List</li>
<li>Circular Singly Linked List</li>
<li>Circular Doubly Linked List</li>
</ul>

There will be an in-depth look on each type in their respective notebooks. Every linked list is composed of nodes pointing to one another. The first node in a linked list will have a pointer called "head" which will always point to the first node of the list.

# Access

Unlike an array, the computer does not perform some calculation to reach the desired index. Instead we start at the head, and we traverse the list until we reach our destination. For instance, if I wanted the data at position 42, we would start at index 0, and follow our pointers until we reach position 42, then return the data. As a result of this process of traversing,<b> access of a linked list is</b> $O(n)$.

# Search

Similar to access, in order to search for an element we must traverse our linked list beginning from the head until we find what ever piece of data we are looking for. In the worst case, we may never find our data and we traverse the entire list to the $nth$ item. <b>Therefore search of a linked list is</b> $O(n)$.

# Insertion

Insertion is a three step procedure:

1) Find the index of insertion

Let's say we want to insert $B$ inbetween $A$ and $C$. Once we find our index of insertion, our next step is:

2) Have the new node point to the next node

```python
from IPython.display import Image

Image("images/ins2.png")
```

$B$ is now pointing to $C$. Our final step is:

3) Have the previous node point to our new node

```python
from IPython.display import Image

Image("images/ins3.png")
```

$A$ now points to $B$ and $B$ has been successfully inserted.
<hr>

The time complexity of insertion is $O(1)$ because the act of reassigning pointers is just an assignment statement which operates in constant time. <b>It is important to note that we assume that we are already at the index of insertion</b>. We assume this because we only care about the time complexity of the actual act of insertion, the reassignment of pointers, not including the time it takes to find the index of insertion. If we account for finding the index, then insertion would be $O(n)$. However, the industry always uses $O(1)$ as the time complexity of insertion.
