# Queue Code Breakdown

The queue is an abstract data structure that implements a First-in First-out(FIFO) policy. Where the first piece of data that enters the data structure, will be the first piece of that data that exits that data structure. The queue can be implemented via other concrete data structures, such as arrays, and linked lists.

From there I built standard queue algorithms. size() returns the length counter, isFull() checks if the queue is full, isEmpty() checks if the queue is empty.

```python
public int size()
    {
        return length;
    }
    public boolean isFull()
    {
        return theQueue.size() == length;
    }

    public boolean isEmpty()
    {
        return theQueue.size() == 0;
    }
```

The queue's main operations are enqueue(), dequeue() and peek(). enqueue() will add items into the queue, if the queue is full, the user will recieve a queue overflow error.

```python
 public void enqueue(int data)
    {
        if(isFull())
        {
            System.out.println("The queue is full");
            return;
        }
        theQueue.add(data);
    }
```

dequeue() will remove the first item on the queue, if the queue is empty, the user will recieve a queue underflow error.

```python
 public int dequeue()
    {
        if(isEmpty())
        {
            System.out.println("The queue is empty");
            return;
        }
        return theQueue.remove(0);
    }
```

peek() will let the user see the first item on the queue, but it will not remove it.


```python
public int peek()
    {
        if(isEmpty())
        {
            System.out.println("The queue is empty");
            return -1;
        }
        return theQueue.get(0);
    }
```

Below is my Queue class in its entirety.

```python
package queue;

import java.util.*;

public class Queue {

    private List<Integer> theQueue = new ArrayList<Integer>();
    private int length;

    public Queue(int length)
    {
        this.length = length;
    }

    public int size()
    {
        return length;
    }
    public boolean isFull()
    {
        return theQueue.size() == length;
    }

    public boolean isEmpty()
    {
        return theQueue.size() == 0;
    }

    public void enqueue(int data)
    {
        if(isFull())
        {
            System.out.println("The queue is full");
            return;
        }
        theQueue.add(data);
    }

    public int dequeue()
    {
        if(isEmpty())
        {
            System.out.println("The queue is empty");
            return;
        }
        return theQueue.remove(0);
    }

    public int peek()
    {
        if(isEmpty())
        {
            System.out.println("The queue is empty");
            return -1;
        }
        return theQueue.get(0);
    }
}
```
