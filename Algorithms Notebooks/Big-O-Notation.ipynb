# Big O Notation


Formally known as <b>Asymptotic Analysis</b>, the purpose of this topic is to define the mathematical boundation/framing of an algorithm's run-time performance. Through asymptotic analysis, we are able to determine the
<ul style="list-style-type:disc">
  <li><b> Best Case (Ω Notation)</b></li>
  <li><b> Average Case (θ Notation)</b></li>
  <li><b> Worst Case (Ο Notation)</b></li>
</ul>
of an algorithm. 

# Execution Time Cases

<b> Best Case </b> - Least possible execution time of operation.<br>
<b> Average Case </b> - Average execution of operations. If an operation takes $f(n)$ time, then the average case will take less than $f(n)$.<br>
<b> Worst Case </b> - Algorithm takes maximum time to execute, $f(n)$.<br>
<hr>

In the software industry, we are primarily concerned with the worst case, and ocassionally the average case; however we will mostly study the worst case of any algorithm. Below is a reference to common run-time performances:

# Determing Run-Time

<b> Assignment operations</b> always operate at $O$(1) time (pronounced as "order of one") because the computer just creates a variable that will point to some value.

```scijava
int a = 5;
String s = "Hello";
String t = "World!";
```

<b>Arithmetic operations</b> always operate at $O$(1) time (the reason why is a whole other topic).

```scijava
int a = 5;
int b = 7;
int c = b + a;
System.out.println(c - b * b/a);
```

<b>Iterations</b> such as a for loop operate at $O$($n$) ("order of $n$") time because we are traversing through our data set $n$ times.

```scijava
for(int i = 0; i < n; i++){}
```

It should be noted that you can determine the run time of a for loop by its terminating condition $i < n$; The condition may not use $n$. For instance, $i < array.length;$ in this case the loop operates at $O$($array.length$) time what ever that number may be. But we may let $n$ represent $array.length$, so it may be generalized as $O$($n$).

<b>Nested Iterations</b> such as a two layered for loop is slightly more complicated in terms of determining its run-time. This is because our nested for loops may have different terminating conditions such as below:

```scijava
for(int i = 0: i < n: i++)
{
    for(int j = 0; j < m: j++)
    {
        
    }
}
```

You cannot assume $n = m$, therefore our run-time is $O$($nm$) ("order of $nm$"). For every value of $n$, we must do something $m$ times: $n\times m$. On the other hand if our nested for loops have the same terminating condition such as below:

```scijava
for(int i = 0: i < n: i++)
{
    for(int j = 0; j < n: j++)
    {
        
    }
}
```

We can conclude the run-time is $O$($n^2$) ("order of $n$ squared"). For every value of $n$, we must do something $n$ times: $n\times n$. We may even have many more layers of loops, if they all have the same terminating condition we can say the algorithm's run-time is $O$($n^m$) ("order of $n$ raised to $m$") where $m$ is the number of nested for loops.

```scijava
for(int i = 0: i < n: i++)
{
    for(int j = 0; j < n: j++)
    {
        for(int k = 0; k < n; k++)
        {
            for(int l = 0; k < n; k++)
            {
                for(int m = 0; k < n; k++)
                {
                    //O(n^5) time
                }   
            }   
        }
    }
}
```

# Analyzing an Algorithm

An algorithm typically compromises of more than just one assignment statement, or one for loop. There are many operations occurring which we can calculate the the run-time of each, then sum those times to get our total run-time for the entire algorithm. This is our function of time:<br>
$$T(n) = \sum_{i=1}^{n} O(n_i)$$ <br>
Where $n_i$ may be any variation of a run-time ($n^2$, $n!$, $n$, 1, $log (n)$).<br> <b>It is critical to remember that since we are trying to computer the worst case of an algorithm we only keep the dominant term in our function of time.</b>

# Example 1

```scijava
int k = 0;//Assignment operation: O(1)
for (int i = 1; i <= n; i++)//Iteration n times: O(n) 
{
    k = k + 5;//Arithmetic operation: O(1)
}
```

For every iteration of the loop we perform $k = k + 5$ $n$ times: $O(n*1)$
Our function of time composes of:
$$T(n) = O(1) + O(n) * O(1)$$
$$T(n) = O(1) + O(n*1)$$ 
$$T(n) = O(n)$$<br>
Since $O(n)$ is our dominant term, when we add $O(1) + O(n*1)$ we only keep $O(n)$. This example has a run-time of $O(n)$.

# Example 2

```scijava
int k = 0;//O(1)
for (int i = 1; i <= n; i++)//O(n) 
{
    for (int j = 1; j <= n; j++)//O(n)
    {
        k = k + i + j;//O(1)
    }
}
```

Let's break this down. The outer loop executes $n$ times and for every $n$ we will execute the inner loop $n$ times. Thus, the time complexity for this snippet of code is:
$$T(n) = O(1) + O(n) * O(n) * O(1)$$
$$T(n) = O(1) + O(n) * O(n*1)$$
$$T(n) = O(1) + O(n*n*1)$$
$$T(n) = O(1) + O(n^2)$$
$$T(n) = O(n^2)$$<br>

# Example 3

```scijava
int k = 0;// O(1)
for (int i = 1; i <= n; i++)//O(n) 
{
    for (int j = 1; j <= i; j++)//?? 
    {
        k = k + i + j;//O(1)
    }
}
```

The outer loop executes n times. For i = 1, 2,..., the inner loop is executed one time,
two times, and n times. Let c represent a constant. Thus, the time complexity for the loop is:
$$T(n) = O(c) + O(2c) + O(3c) + O(4c) + . . . + O(nc)$$
$$T(n) = O(cn)*O(\frac{n+1}{2})$$
$$T(n) = O(\frac{cn^2}{2}) + O(\frac{cn}{2})$$
$$T(n) = O(n^2)$$
