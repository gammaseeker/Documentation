Today I attended a product demo. The product was designed to reduce the friction in the development process when making changes to code by automizing some of the nuances during development. The demo demonstrated features such as auto-write ups during documentation, a new hot fix feature in the work flow, and more technical aspects that I didn't really understand. But the one thing I took away from this event was the four stages of software development and the purpose of GitHub branches:
<ul>
    <li> 
        Alpha: Sandbox developer mode
    </li>
    <li>
       Beta: Internal testing mode
    </li>
    <li>
       Beta 2: Internal testing mode
    </li>
    <li>
        Prod/Prodction: Final Testing before deployment
    </li>
    <li>
        Good habit: master branch should always be in a deployable state
    </li>
    <li>
        Branches are for organization purposes, you don't want your edits to affect other people's edits and vice versa
    </li>
</ul>
