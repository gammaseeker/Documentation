# Golf Anecdote

Mike hit on a point where if you want to be successful in the world, you're going to need more than just the technical skills. You need the personality and work ethic to suceed. He explained this point via an anecdote. He went golfing and there was this 15 year old boy working at the range. He stood out to Mike because the kid would consistently offer help to the patrons even though his job was to pick up golf balls on the field. He said things like "I'm not an expert on golfing but this form would help...", "Would you like any refreshments..." etc. I quote Mike, "If it weren't for child labor laws, I would've snatched that kid up to come work here."
