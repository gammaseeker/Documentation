# The MVC Architecture 

Rails uses the Model-view-controller(MVC) architecture in its projects. The model handles the backend work of the application. The model is where all the data mangement, logic, and rules of the application go. The view is the front end component which serves as output to the user after the model finishes executing. The controller takes input from the user and provides commands to either the model or view.
